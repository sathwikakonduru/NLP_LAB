```python
import nltk
from nltk.util import ngrams

text = "I love natural language processing".split()
bigrams = list(ngrams(text, 2))
trigrams = list(ngrams(text, 3))

print("Bigrams:", bigrams)
print("Trigrams:", trigrams)
```
