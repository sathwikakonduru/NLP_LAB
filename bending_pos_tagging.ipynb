```python
import spacy
nlp = spacy.load("en_core_web_sm")

text = "The striped bats are hanging on their feet for best"
doc = nlp(text)

for token in doc:
    print(f"{token.text:<10} | {token.lemma_:<10} | {token.pos_}")
```
