```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

# Download necessary NLTK data
nltk_packages = ['punkt', 'wordnet', 'averaged_perceptron_tagger', 'averaged_perceptron_tagger_eng', 'punkt_tab']
for package in nltk_packages:
    try:
        nltk.data.find(f'tokenizers/{package}')
    except LookupError:
        nltk.download(package)


text = "The striped bats are hanging on their feet for best"
lemmatizer = WordNetLemmatizer()

def pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    return {'J': wordnet.ADJ, 'N': wordnet.NOUN, 'V': wordnet.VERB, 'R': wordnet.ADV}.get(tag, wordnet.NOUN)

words = nltk.word_tokenize(text)
print("Lemmatized Text:", " ".join([lemmatizer.lemmatize(w, pos(w)) for w in words]))
```
