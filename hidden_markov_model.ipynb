```python
# Hidden states
states = ['Rainy', 'Sunny']

# Observations
observations = ['walk', 'shop', 'clean']

# Start probabilities
start_prob = {'Rainy': 0.6, 'Sunny': 0.4}

# Transition probabilities
trans_prob = {
    'Rainy': {'Rainy': 0.7, 'Sunny': 0.3},
    'Sunny': {'Rainy': 0.4, 'Sunny': 0.6}
}

# Emission probabilities
emit_prob = {
    'Rainy': {'walk': 0.1, 'shop': 0.4, 'clean': 0.5},
    'Sunny': {'walk': 0.6, 'shop': 0.3, 'clean': 0.1}
}

# Example observation sequence
obs_seq = ['walk', 'shop', 'clean']

# Simple forward algorithm to calculate sequence probability
prob = 1.0
prev_state = None
for i, obs in enumerate(obs_seq):
    if i == 0:
        prob *= sum(start_prob[s] * emit_prob[s][obs] for s in states)
    else:
        prob *= sum(trans_prob[prev_state][s] * emit_prob[s][obs] for s in states)
    prev_state = 'Rainy'  # just pick one path for simplicity

print("Approx sequence probability:", prob)
```
